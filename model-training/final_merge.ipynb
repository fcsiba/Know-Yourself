```python
import pandas as pd
import numpy as np
import pickle
```

```python
original_csv_data = open('dataset\\original_dataset.csv', 'r')

original_dataframe = pd.read_csv(original_csv_data)
```

```python
original_tweets = original_dataframe['content'].values
```

```python
file = open('all_classes101.pkl', 'rb')
original_classes = pickle.load(file)
file.close()
```

```python
file = open('angry_tweets101.pkl', 'rb')
extra_anger_tweets = pickle.load(file)
file.close()

file2 = open('anger_class101.pkl', 'rb')
extra_anger_classes = pickle.load(file2)
file2.close()

extra_anger_tweets = extra_anger_tweets[0:8000]
extra_anger_classes = extra_anger_classes[0:8000]
```

```python
file = open('hate_tweets101.pkl', 'rb')
extra_hate_tweets = pickle.load(file)
file.close()

file2 = open('hate_class101.pkl', 'rb')
extra_hate_classes = pickle.load(file2)
file2.close()
```

```python
final_all_tweets = np.concatenate((original_tweets, extra_anger_tweets, extra_hate_tweets))
```

```python
final_all_classes = np.concatenate((original_classes, extra_anger_classes, extra_hate_classes))
```

```python
d = {'tweet':final_all_tweets,'class':final_all_classes}
final_dataframe = pd.DataFrame(d)
```

```python
unique_sentiments = final_dataframe['class'].unique().tolist()
unique_sentiments
```

```python
final_dataframe_shuffled = final_dataframe.iloc[np.random.permutation(final_dataframe.index)].reset_index(drop=True)
final_dataframe_shuffled
```

```python
with open('final_all_data2.pkl', 'wb') as f:
   pickle.dump(final_dataframe_shuffled, f)
```
